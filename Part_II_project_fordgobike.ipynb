```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
fordgobike_clean = pd.read_csv('fordgobike_clean.csv')
fordgobike_clean.sample(5)
```

## (Does the bike share trip depends on if the user type is costumer or subscriber)

> This is a multiple planatory visualiztion that shows the relationship between user_type, bike_share_ for_all_trip 
the San francisco bar share trip has  158386 Subscriber and 16566 Customer       


```python
#plotting a cluster bar chart to show the relationship between bike share for all trip and user type
plt.figure(figsize = [10,8])
sb.countplot(data = fordgobike_clean, x = 'bike_share_for_all_trip', hue = 'user_type')
plt.xticks(rotation = 0)
plt.ylabel('user type count')
plt.xlabel('Bike share for all trip option')
plt.title('Clustter bar chart of user type and bike share for all trip in SF Bike Share');
```

## (Does gender affect the bike share trip for riders in the Sans Francisco)

>Majority of the riders are of the Age bracket of 18 to 45 years. and the male is 130500, female is 40805 while other is 3647

```python
# plotting a facetgrid graph of member gender, duration time in seconds and age 
genders = sb.FacetGrid(data = fordgobike_clean, col = 'member_gender',col_wrap = 3, size = 5)
genders.map(plt.scatter, 'duration_sec', 'Age',alpha = 0.25)
genders.fig.set_size_inches(14,8);
plt.ylabel('Age (years)')
plt.xlabel('Duration time (sec)')
plt.suptitle('Age, member gender and duration time in San Francisco Bike Share',y = 1, fontsize = 14, weight = "bold");
plt.show()
```

## (How does age depend on how fast the riders finish the trip?)
> The highest duration time were from riders with about 18 to 45years of age 


```python
#plotting a heap map of age against duration in second
plt.figure(figsize = [8, 6])
plt.hist2d(data = fordgobike_clean, x = 'duration_sec', y = 'Age', cmin = 0.5, cmap = 'viridis_r');
plt.colorbar()
plt.ylabel('Age(years)')
plt.xlabel(' duration seconds')
plt.title('heap map correlation between Duration and Age in San Francisco Bike Share trip');
plt.show()
```
